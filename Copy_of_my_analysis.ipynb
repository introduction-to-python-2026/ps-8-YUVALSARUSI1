```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes

# ============================
# 1. LOAD DATA (always available)
# ============================

data = load_diabetes()
df = pd.DataFrame(data.data, columns=data.feature_names)
df["target"] = data.target

print("Data loaded successfully!")
print(df.head())

# ============================
# 2. SAVE RAW FILE LOCALLY
# ============================

df.to_csv("diabetes_raw.csv", index=False)
print("\nRaw file saved as diabetes_raw.csv")

# ============================
# 3. BASIC DATA CLEANING
# ============================

print("\nMissing values before cleaning:")
print(df.isnull().sum())

# This dataset has no missing values, but we'll pretend to clean:
df_clean = df.copy()

# Columns to lowercase (example preprocessing)
df_clean.columns = df_clean.columns.str.lower()

print("\nMissing values after cleaning:")
print(df_clean.isnull().sum())

# ============================
# 4. VISUALIZATION
# ============================

plt.figure(figsize=(8, 6))
plt.scatter(df_clean["bmi"], df_clean["target"])
plt.title("Diabetes Disease Progression vs BMI")
plt.xlabel("BMI")
plt.ylabel("Disease Progression Score")
plt.show()
```
